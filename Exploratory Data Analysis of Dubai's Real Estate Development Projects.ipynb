```python
# Replace all VALUE errors
df.replace(["#VALUE", "#VALUE!"], np.nan, inplace=True)

# Fill categorical nulls
df["Zoning Authority"].fillna("Unknown", inplace=True)

# Fix project status spelling
df["Project Status"] = df["Project Status"].replace({"FRIEZED": "FREEZED"})
```

**Handling Invalid Values and Standardizing Categorical Data**

- **Replaced Invalid Placeholder Values**
  - Identified entries such as `#VALUE` and `#VALUE!` in the dataset
  - Converted them to **NaN** to enable proper handling of missing data  

- **Handled Missing Values in Categorical Columns**
  - Filled missing values in the `zoning_authority` column with **"Unknown"**
  - Ensures consistency and prevents errors during grouping and analysis  

- **Corrected Inconsistent Category Labels**
  - Fixed misspelled values in the `project_status` column  
  - Example: `FRIEZED` → `FREEZED`  
  - Helps avoid duplicate categories and improves analysis accuracy  

```python
# Converting the Project Start Date and Project End Date columns into datetime format for better date analysis.
df["Project Start Date"] = pd.to_datetime(df["Project Start Date"], errors="coerce",dayfirst=True)
df["Project End Date"] = pd.to_datetime(df["Project End Date"], errors="coerce",dayfirst=True)
df.dtypes
```

The project start and end date is converted to datetime for better analysis.

```python
# Drop rows with missing start date
df.dropna(subset=["Project Start Date"], inplace=True)

# Fill Master Project
df["Master Project"].fillna("Unknown", inplace=True)

# Fill Percent Completed
df["Percent Completed"].fillna(0, inplace=True)

# Fill missing Project End Date with today's date
df["Project End Date"].fillna(pd.Timestamp.today(), inplace=True)

# Create Project Duration
df["Project Duration (Days)"] = (df["Project End Date"] - df["Project Start Date"]).dt.days
```

```python
# Create subplots
fig = make_subplots(rows=2, cols=2,subplot_titles=("<b>Project Duration Distribution</b>","<b>Completion Percentage Distribution</b>","<b>Projects Started Over Years</b>",
                                                   "<b>Project Duration by Type </b>"),horizontal_spacing=0.15,vertical_spacing=0.15)

# 1️⃣ Histogram - Project Duration
fig.add_trace(go.Histogram(x=df['project_duration_(days)'],nbinsx=30,marker=dict(color='royalblue')),row=1, col=1)

# 2️⃣ Bar Plot- completion categories
bins = [0, 25, 50, 75, 100]
labels = ['0-25%', '25-50%', '50-75%', '75-100%']
df['completion_category'] = pd.cut(df['percent_completed'], bins=bins, labels=labels, include_lowest=True)
completion_counts = df['completion_category'].value_counts().sort_index()
fig.add_trace(go.Bar(x=completion_counts.index,y=completion_counts.values,marker=dict(color='seagreen')),row=1, col=2)

# 3️⃣ Line Plot - Projects per Year
yearly = df['start_year'].value_counts().sort_index()
fig.add_trace(go.Scatter(x=yearly.index,y=yearly.values,mode='lines+markers',line=dict(color='orange', width=3)),row=2, col=1)

# 4️⃣ Box Plot - Duration by Project Type
for ptype in df['project_type'].dropna().unique():
    fig.add_trace(go.Box(y=df[df['project_type'] == ptype]['project_duration_(days)'],name=ptype,boxmean=True),row=2, col=2)

# Axis labels
fig.update_xaxes(title_text="Project Duration (Days)", row=1, col=1)
fig.update_yaxes(title_text="Number of Projects", row=1, col=1)

fig.update_xaxes(title_text="Completion Category", row=1, col=2)
fig.update_yaxes(title_text="Number of Projects", row=1, col=2)

fig.update_xaxes(title_text="Year", row=2, col=1)
fig.update_yaxes(title_text="Number of Projects", row=2, col=1)

fig.update_xaxes(title_text="Project Type", row=2, col=2)
fig.update_yaxes(title_text="Project Duration (Days)", row=2, col=2)

# Layout
fig.update_layout(height=800,width=1000,title_text="<b>📊 Dubai Real Estate Insights Dashboard</b>",title_x=0.5,template="plotly_white",showlegend=False)
fig.show()
```
